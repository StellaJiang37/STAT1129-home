### a.

```r
Data_Frame <- data.frame(
    id = c(1,2,3,4,5),
    name =c("Peter", "Amy", "Ryan", "Gary", "Michelle"),
    salary =c(623.30, 515.20, 611.00, 729.00, 843.25)
) 
Data_Frame
```

### b.

```r
n_Data_Frame<- cbind(Data_Frame, department=c("IT", "fiance", "CS", "literature", "statistics"))
n_Data_Frame
```

### c.

```r
n_Data_Frame[c(1,3,5), c(2,3)]
```

### d.

```r
Data_Frame[c(1,4,5), c(3)]
Data_Frame[c(1,4,5),c(2)]
```

```r
x<- Data_Frame[c(1,4,5),c(2)]
y<- Data_Frame[c(1,4,5), c(3)]

barplot(y, names.arg=x)
                
```

### e.

```r
highest<- max(Data_Frame[,c(3)])
highest
```

```r
lowest<- min(Data_Frame[,c(3)])
lowest
```

```r
median<- median(Data_Frame[,c(3)])
median
```

```r
v1<- c(highest, lowest, median)
v1
v2<- c('highest salary', 'lowest salary', 'median salart')
v2
```

```r
pie(v1, label=v2, main="Analysis of salary")
```
