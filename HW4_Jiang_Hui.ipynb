## Question5

```python
marks={'Andy':88, 'Amy':66, 'James':90, 'Jules':55, 'Arthur':77}
grades = marks.values()
grades
```

```python
name=marks.keys()
name
```

```python
import matplotlib.pyplot as plt


mylabels=name
plt.pie(grades, labels=mylabels)
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')

plt.title("GRADES")
plt.show()
```

```python
plt.bar(name,grades)
plt.title("GRADES")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#multiple lines
y1=np.array([3,9,1,10])
y2=np.array([6,2,7,11])
plt.subplot(2,3,1)
plt.plot(y1)
plt.plot(y2)
plt.title('Multiple lines')

#bar chart
plt.subplot(2,3,2)
plt.bar(name,grades)
plt.title("Bar chart")

#pie chart

plt.subplot(2,3,3)
plt.title("Pie Chart")
plt.pie(grades, labels=mylabels)


# grid
x=np.linspace(80,125,10)
y=np.linspace(240,330,10)
plt.subplot(2,3,4)
plt.plot(x,y)
plt.title("Plot with grid")
plt.grid()

#scatter
x=np.array([1,5,9,2,4,3,5,11,3,15])
y=np.array([12,9,3,16,2,5,1,8,10,2])
plt.subplot(2,3,5)
plt.scatter(x,y)
plt.title("Scatter")

#histogram
mu, sigma=0,0.1
s=np.random.normal(mu, sigma, 1000)
plt.subplot(2,3,6)
plt.hist(s)   
plt.title("Histogram")


plt.subplots_adjust(hspace=0.35)
plt.suptitle("Subplot Pratice")
plt.show()
```
